```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Salary Data.csv')
```

```python
df.head()
```

```python
# Checking for missing values
print(df.isnull().sum())
# Print the cleaned dataframe info
print(df.info())
```

```python
df["Education Level"].unique()
```

```python
df["Job Title"].unique()
```

```python
age_range = (df["Age"].min(), df["Age"].max())
print("Age Range:", age_range)
```

```python
import matplotlib.pyplot as plt
# Count the frequency of each age
age_counts = df["Age"].value_counts().sort_index()

# Plotting the line chart
plt.figure(figsize=(10, 5))
plt.plot(age_counts.index, age_counts.values, marker='o', linestyle='-')
plt.title("Age Distribution Line Chart")
plt.xlabel("Age")
plt.ylabel("Number of Employees")
plt.grid(True)
plt.show()
```

```python
yoe_range = (df["Years of Experience"].min(), df["Years of Experience"].max())
print("Years of Experience Range:", yoe_range)
```

```python
# Count the frequency of each yoe
yoe_counts = df["Years of Experience"].value_counts().sort_index()

# Plotting the line chart
plt.figure(figsize=(10, 5))
plt.plot(yoe_counts.index, yoe_counts.values, marker='o', linestyle='-')
plt.title("Years of Experience Line Chart")
plt.xlabel("Years of Experience")
plt.ylabel("Number of Employees")
plt.grid(True)
plt.show()
```
